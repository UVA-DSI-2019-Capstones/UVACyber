```python
import pandas as pd
```

```python
# read in honeypot2 csv as a pandas df
df = pd.read_csv('honeypot2_1.csv', encoding='utf-8')
```

```python
df
```

```python
# create flag variable for connections starts and end 
res= []
for index,row in df.iterrows():
    if "New connection" in row['details']:
        res.append(1)
    elif "Connection lost after" in row['details']:
        res.append(2) 
    else:
        res.append(0)
df['flag']=res
```

```python
df
```

```python
# sort df by source IP and then timestamp 
df = df.sort_values(by=['src_ip', 'timestamp'])
```

```python
df
```

```python
# subset if flag column is equal to 0
df = df[df["flag"]!=0]
```

```python
# export csv 
df.to_csv("collision.csv",header = True, index= True, encoding = "utf-8")
```
