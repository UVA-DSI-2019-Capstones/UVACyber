```python
# read in honeypot1 csv as a pandas df
df = pd.read_csv('honeypot1_1.csv', encoding='utf-8', index_col=None)
```

```python
df
```

```python
# create flag variable for connections starts and end 
res= []
for index,row in df.iterrows():
    if "New connection" in row['details']:
        res.append(1)
    elif "Connection lost after" in row['details']:
        res.append(2) 
    else:
        res.append(0)
df['flag']=res
```

```python
df
```

```python
# strip IPs of whitespace 
df['src_ip']=[x.lstrip() for x in df['src_ip']]
df['src_ip']=[x.split('.') for x in df['src_ip']]
df['src_ip'] = [''.join(x) for x in df['src_ip']]
```

```python

df['src_ip'].tail()
```

```python
# sort df by source IP and then timestamp 
df = df.sort_values(by=['src_ip', 'timestamp'])
```

```python
df
```

```python
# subset if flag column is equal to 0
df = df[df["flag"]!=0]
```

```python
df
```

```python
# get timestamps of collisions
collision_timestamp=[]
previous='2'
for index,row in df.iterrows():
    if previous==row['flag']:
        collision_timestamp.append(row['timestamp'])
    previous=row['flag']
```

```python
# percentage of data (collisions) removed
len(collision_timestamp)/float(len(df))
```

```python
# subset data by these timestamps 
res=df.loc[~df['timestamp'].isin(collision_timestamp)]
```

```python
len(res)
```

```python
# export csv 
res.to_csv("collision.csv",header = True, index= False, encoding = "utf-8")
```
